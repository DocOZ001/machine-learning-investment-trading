```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from matplotlib import style
style.use("ggplot")

X = np.array([[1,2],[5,8],[1.5,1.8],[8,8],[1,0.6],[9,11]])
y = [0,1,0,1,0,1]
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
# Now let's train an svm with these samples and plot the decision boundary
```

```python
clf = svm.SVC(kernel = 'linear')
clf.fit(X,y)
print(clf.predict([10.58,10.76]))
```

```python
# The new point is correctly classified as [1]
plt.scatter(X[:,0],X[:,1],c=y)
plt.scatter(10.58,10.76,c='k',marker='o',linewidths=10) 
```

```python
w = clf.coef_[0]
print('Feature Weights',w)
a = -w[0] / w[1]
print(a)
# let's get the end points of the decision line
# xx = np.linspace(0,12) in the video, but end points are sufficient
xx = np.array([0,12])
yy = a * xx - clf.intercept_[0] / w[1]
print(xx)
print(yy)
plt.plot(xx,yy,label='Decision Boundary')

plt.scatter(X[:,0],X[:,1],c=y)
plt.scatter(10.58,10.76,c='k',marker='o',linewidths=10) 
plt.legend(loc=2)
```
